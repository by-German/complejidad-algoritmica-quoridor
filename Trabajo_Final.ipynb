# Backtracking
El Backtracking es una técnica utilizada en la programación para llevar a cabo una búsqueda sistemática por medio de todas las variables posibles en un espacio de búsqueda. Los algoritmos de Backtracking se dedican a la construcción de soluciones posibles de manera que se verifica si esta variable cumple con los requisitos de la solución o no, y con esta información se continúa con el proceso de busqueda. De esta forma, se construyen todas los posibles caminos de la solución , y se invoca recursivamente al algoritmo en las mismas.

Una breve explicación del comportamiento del backtracking es decir que consiste en enumerar de manera sistemática las alternativas de solución al problema y al probar cada una de las alternativas se guarda en memoria del resto de alternativas, esto servirá en el caso de que no se elija el camino correcto para llegar a la solución, se pueda volver a un punto para probar con otras alternativas.
 
Esta técnica es usualmente representada a modo de árbol debido a que facilita el entendimiento del mismo. El árbol se recorre en profundidad. Un camino a la solución está dada por un conjunto de vértices de tal forma que se elabora una secuencia que logra resolver el problema.

### Gerneración del Grafo:
El Quoridor, es un tablero de dimension $ 9 \times 9$, en donde los jugadores pueden desplzarse normalmente hacia arriba / abajo o izquierda / derecha, mientras tengan la posibilidad de hacerlo. Por ello, para representarlo se ha optado por hacerlo mediante la generacion de un grafo, ya que de este modo se podra simular los posibles movimientos que podra hacer el jugador en todo el partido a traves de las aristas de grafo.

A continuacón, se representa la funcion que nos permitira generar el grafo, con dimensiones arbitrarias $ n \times n $:

```
def grafo_tablero(n):   # Complejidad -> O(n^2)
    G = nx.path_graph(n)
    G.remove_node(0)
    count = 0
    for x in range(n):
        for y in range(n):
            count += 1
            if not (y == n - 1): G.add_edge(count, count + 1)
            if not (x == n - 1): G.add_edge(count + n, count)
            G.nodes[count]["id"] = count
    return G
```

```
n = 5
G = grafo_tablero(n)
nx.draw(G, with_labels = True)
```

#### Implementacion del algoritmo BFS:

```
def BFS(G, s):
    for _, u in G.nodes(data = True):
        u["color"] = "blanco"
        u["distancia"] = None # infinito
        u["padre"] = None
    s["color"] = "gris"
    s["distancia"] = 0
    s["padre"] = None
    Q = deque()
    Q.append(s) # enqueue
    while not (Q == deque([])):
        u = Q.popleft()
        for _, v in G.edges(u["id"]):
            nodo = G.nodes[v]
            if nodo["color"] == "blanco":
                nodo["color"] = "gris"
                nodo["distancia"] = u["distancia"] + 1
                nodo["padre"] = u # u["id"]
                Q.append(nodo)
        u["color"] = "negro"
```

#### Implementacion del algoritmo DFS:

```
def DFS(G, s):
    for id, node in G.nodes(data = True): # node is atribute
        node['id'] = id
        node['vis'] = False
        node['padre'] = None

    DFS_visit(G, s)

def DFS_visit(G, node):
    node['vis'] = True
    for id in G.adj[node['id']]: # vecinos
        if G.nodes[id]['vis'] == False:
            G.nodes[id]['padre'] = node
            DFS_visit(G, G.nodes[id])
```

#### Implementacion del Algoritmo Djsktra:

```
def Dijkstra(Grafo, inicio, meta):
    shrt = {}  # Grabar el costo para llegar al nodo, shrt de shortpath
    nddesc = Grafo  # itera el grafo completo
    inf = 999999  # Solamente un numero grande, que sera mas grande que los pesos de los nodos
    path = []  # grabar el camino de regreso al nodo, el mas optimo

    for node in nddesc:
        shrt[node] = inf
    shrt[inicio] = 0

    while nddesc:
        min_dist = None
        for node in nddesc:
            if min_dist is None:
                min_dist = node
            elif shrt[node] < shrt[min_dist]:
                min_dist = node

        opciones = Grafo.edges(min_dist, data = "peso")
        for origen,hijos, peso in opciones:
            if peso + shrt[min_dist] < shrt[hijos]:
                shrt[hijos] = peso + shrt[min_dist]
        nddesc.remove_node(min_dist)
    NodoActual = meta
    path.insert(0, inicio)
    if shrt[meta] != inf:
        return str(shrt[meta])
```

#### Implementacion Print-Path:
Esta funcion nos permite conocer cuales son los nodos por los cuales se deben de recorrer , para llegar de un nodo origen $s$ a un nodo destino $v$, segun el algoritmo de ***busqueda de grafos*** utilizado.

```
def hallar_camino(G, s, v, camino):
    if v["id"] == s["id"]:
        camino.append(s["id"])
    elif v["padre"] == None:
        return 0
    else:
        hallar_camino(G, s, v["padre"], camino)
        camino.append(v["id"])
```

Como se puede evidenciar, al remover la arista $(8, 13)$, es símil a colocar un muro entre en casiller $8$ y $13$. Esto representa que si el jugador se encuentra en el casillero $8$ no podra avanzar hacia el casillero $13$ directamente.

Por otro lado, el jugador en el juego debe elegir entre colocar un "muro" o seguir con su camino, en su turno correspondiente. Para ello, se debe de tener en cuenta si el jugador se encuentra perdiendo o ganando respecto a sus adversarios, que como máximo pueden ser 4. En caso el jugador se encuentre ganando, obviamente escogerá seguir con su camino, para llegar a la meta más antes que sus adversarios, sin embargo, al estar perdiendo, tendrá que obstaculizar el camino de sus adversarios, para tener opotunidad de ganar. 
Para ello, es necesario conocer si un jugador esta perdiendo:
Al contar con una ***lista "camino"***  que contiene los nodos del camino de los jugadores, a través de ella, podemos determinar el estado de los jugadores, siendo $j$ el jugador e $i$ los demás jugadores:


$$ jugador_j = 
\left\{
\begin{array}{ll}
len(camino)_j & < & len(camino)_i & ;wining \\
len(camino)_j & > & len(camino)_i & ;losing
\end{array}
\right.
$$

Esto quiere decir, que el jugador estará gananndo si el tamaño de su camino para llegar a su destino es menor que el de los demás jugadores, por lo cual le conviene avanzar. Caso contrario, el jugador decide obstaculizar al resto de los jugadores, impidiendo el avance del jugador que está ganando, colocandole barreras en su camino.


### Algorimo para encontrar **Arista de Corte**:

Para ello, se utilizará una de las funcionalidades de DFS:

```
import numpy as np
```

```
def dfs_bridge(G, time = 0, bridges = []):
    for id, node in G.nodes(data = True):
        node['id'] = id
        node['vis'] = False
        node['p'] = None
        node['low'] = np.inf
        node['disc'] = np.inf
    for id, node in G.nodes(data = True):
        u = G.nodes[id]
        if u['vis'] == False:
            DFS_visit_bridge(G, u, time, bridges)
    return bridges

def DFS_visit_bridge(G, u, time, bridges):
    u['vis'] = True
    u['disc'] = time
    u['low'] = time
    time += 1
    for id in G.adj[u['id']]: # id de los vecinos
        v = G.nodes[id]
        if v['vis'] == False:
            v['p'] = u
            DFS_visit_bridge(G, v, time, bridges)
            # verificar si 'v = G.nodes[id]' tiene una conexion a un ancestro de 'u = node'
            u['low'] = min(u['low'], v['low'])
            if v['low'] > u['disc']:
                bridges.append((u['id'], v['id']))
                bridges.append((v['id'], u['id']))
        elif v != u['p']:
            u['low'] = min(u['low'], v['disc']) # Actualizar el valor bajo de u para llamadas a funciones principales
```

Como se muestra en el gráfico del grafo anterior, si se remuve la arista $u = 2$ y $v = 5$ se estaría encerrando al jugador que se encuentre dentro de uno de los nodos  $(5, 6, 8, 9, 11, 12)$. Hacer esta operación es sumamente costoso para un algorimo que analice todos las aristas posibles que pueden encerrar al jugaor. Para resolver este dilema, se plantea el uso de encontrar las aristas de corte en el grafo, con una de las funcionalidades que nos provee el algortimo DFS. Para medir la eficiencia de este algoritmo, se medirá el tiempo que se tarda en encontrar todas las aristas de corte en el grafo en comparacion con otro algoritmo, que recorre todo el tablero por cada nodo encontrado.

**Otro Algoritmo**

Este algoritmo consiste en encontar alguna arista que impida al jugador desplazarse de un lugar $u$ a un lugar $v$, para ello, es necesario concocer la ubicacion del jugador, y el lugar a donde quiere dirigirse. En este sentido, podemos ejecutar BFS, desde el origen $u$ a todos los destinos posibles del jugador. Donde, si no se encuentra un camino, entonces, se considera que la arista removida, era una arista de corte.

```
def otro_Algoritmo(G, origen, destinos):
    bridges = []
    for i, _ in G.nodes(data = True):
        for j, _ in G.nodes(data = True):
            if G.has_edge(i, j):
                G.remove_edge(i, j)
                BFS(G, G.nodes[origen])
                for destino in destinos:
                    if hallar_camino(G, G.nodes[origen], G.nodes[destino], []) == 0: # no se encontro camino
                        bridges.append((i, j))
                        break
                G.add_edge(i,j)
    print("Aristas de corte: ", bridges)
```

Para la llamada de este algoritmo, previamenete, se debe indicar la arista que se remueve, luego se ejecuta el algoritmo, y retorna la arista, en caso esta sea un puente en el grafo.

## Medición de tiempos para encontrar las aristas de corte


Para este punto, se generara Grafos de tamaño $n \times n$, con las mismas longitudes que se plantearon para las mediciones de los recorridos de los jugadores. En donde, para cada uno de los grafos, se extraeran aristas $(u, v)$, que generen, por lo menos, una arista de corte en el grafo.

```
def medir_tiempo(n, algoritmo, other):
    G = grafo_tablero(n)
    G.remove_edge(1, 2)     # removiendo aritas, para generar por lo menos una arista de corte
    start = time.time()
    if other:
        algoritmo(G, 1, [1 + n])
    else: print("Aristas de corte: ", algoritmo(G, time = 0, bridges = []))
    end = time.time()
    return end - start
```
