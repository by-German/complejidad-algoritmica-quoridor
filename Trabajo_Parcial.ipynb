##Búsqueda en grafos

En términos generales, un grafo es un conjunto de vértices, también conocidos como nodos y aristas que se conforman a partir de un par de vértices, estos pueden estar orientados o no.
Los algoritmos más esenciales para el desarrollo del proyect en grafos en cuanto a busqueda serían los siguientes:
Algoritmo de búsqueda en anchura (BFS) y Algoritmo de búsqueda en
profundidad (DFS).
##Breadth First Search (BFS)


El algoritmo de búsqueda en anchura, conocido como BFS, permite al usuario recorrer los elementos de un grafo comenzando desde la raíz explorando todos los hijos del vertice. Posteriormente explorará cada uno de los hijos hasta hallar una coincidencia.
##Distributed File System (DFS)

El algoritmo de busqueda de profundidad, tambien llamado DFS permite al usuario recorrer todos los vértices de un grafo, y cuando se llegue al final, recorrer un camino de regreso, con el metodo de Backtracking, de esta forma repetir el mismo proceso con cada hermano del vertice visitado anteriormente.

# Backtracking
El Backtracking es una técnica utilizada en la programación para llevar a cabo una búsqueda sistemática por medio de todas las variables posibles en un espacio de búsqueda. Los algoritmos de Backtracking se dedican a la construcción de soluciones posibles de manera que se verifica si esta variable cumple con los requisitos de la solución o no, y con esta información se continúa con el proceso de busqueda. De esta forma, se construyen todas los posibles caminos de la solución , y se invoca recursivamente al algoritmo en las mismas.

Una breve explicación del comportamiento del backtracking es decir que consiste en enumerar de manera sistemática las alternativas de solución al problema y al probar cada una de las alternativas se guarda en memoria del resto de alternativas, esto servirá en el caso de que no se elija el camino correcto para llegar a la solución, se pueda volver a un punto para probar con otras alternativas.
 
Esta técnica es usualmente representada a modo de árbol debido a que facilita el entendimiento del mismo. El árbol se recorre en profundidad. Un camino a la solución está dada por un conjunto de vértices de tal forma que se elabora una secuencia que logra resolver el problema.
